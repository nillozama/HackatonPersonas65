```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np
import tabula as tb
import requests
import os
import wget
pd.set_option('display.max_rows', 100)
```

```python
datos = pd.read_csv('2019_divter.csv')
datos
```

Descargamos los pdf's con la info de los barrios

```python
urls=datos[datos.CATEGORIA_DIVISIO=='Barri']['URL_FITXA_DIVISIO_TERRITORIAL']
datos_bcn=pd.DataFrame()
for i, url in enumerate(urls):
    file='barri{}.pdf'.format(i+1)
    df= tb.read_pdf(file, pages='all')[0]     #Pasamos la información del pdf a un dataframe
    #Extrae la info
    barri= 'B-{:02d}'.format(i+1)
    poblacion= df.iloc[0,1]
    densidad= df.iloc[2,1]
    poblacion_mas_65= df.iloc[10,1]
    titulados_sup= df.iloc[-10,1]
    poblacion_sola= df.iloc[-8,1]
    indice_sobreenv= df.iloc[-7,1]
    renta_familiar= df.iloc[-2,1]
    #Añadir la info de cada barrio a datos_bcn
    df1=pd.DataFrame({'barri':barri, 'poblacion':poblacion, 'densidad':densidad, 'poblacion_mas_65':poblacion_mas_65, 'titulados_sup':titulados_sup, 'poblacion_sola':poblacion_sola, 'indice_sobreenv':indice_sobreenv, 'renta_familiar':renta_familiar}, index=[0])
    df2.append(df1)
```

```python
df2
```

Generamos el maestro de barrios con el código del barrio, el nombre y el Distrito

```python
barris=datos[datos.CATEGORIA_DIVISIO=='Barri']
barris1= barris.drop_duplicates(subset=['CODI_DIVISIO_TERRITORIAL', 'CODI_DIVISIO_TERRITORIAL_PARE','NOM_DIVISIO_TERRITORIAL'])
barris=barris1[['CODI_DIVISIO_TERRITORIAL', 'CODI_DIVISIO_TERRITORIAL_PARE','NOM_DIVISIO_TERRITORIAL']]
barris
```

Generamos el maestro de Distritos con el código del Distrito y el nombre

```python
districtes=datos[datos.CATEGORIA_DIVISIO=='Districte']
districtes1= districtes.drop_duplicates(subset=['CODI_DIVISIO_TERRITORIAL', 'NOM_DIVISIO_TERRITORIAL'])
districtes=districtes1[['CODI_DIVISIO_TERRITORIAL', 'NOM_DIVISIO_TERRITORIAL']]
districtes=districtes.rename(columns={'CODI_DIVISIO_TERRITORIAL':'Districte' })
districtes
```

Generamos el archivo final con los datos de los barrios y distritos

```python
datos_bcn=df2.merge(barris, how='left', left_on='barri', right_on='CODI_DIVISIO_TERRITORIAL')
```

```python
# Creamos el csv file 
datos_bcn.to_csv('datos_bcn.csv', sep=';', index = False )
```
